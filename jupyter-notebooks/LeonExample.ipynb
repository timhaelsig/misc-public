Import Libraries that are necessary

```python
import numpy
import numpy.random

import matplotlib
import matplotlib.pyplot

import pandas
```

Define or import some variables

```python
# Gaussian random variable with zero mean and standard deviation one, 1000 samples are generated
gaussian_var = numpy.random.normal(0,1,1000)

gaussian_var2 = numpy.random.normal(0,1,1000)

# Multivariate Gaussian, i.e., multiple gaussian random variables with a certain desired correlation
# This example creates four gaussian variables where the second one is uncorrelated to all the others

mean = [0, 0, 0, 0] # vector with mean of each random variable
cov = [[1, 0, 0.9, 0.3], [0, 1, 0, 0], [0.9, 0, 1, 0], [0.3, 0, 0, 1]] # Covariance matrix, describing their relation. If it is diagonal then they are uncorrelated
print(cov)

gaussian_multivar = numpy.random.multivariate_normal(mean, cov, 1000).T
```

Do something with the values

```python
print(numpy.mean(gaussian_var)) # mean of gaussian variable
print(numpy.median(gaussian_var)) # median of gaussian variable
print(numpy.std(gaussian_var)) # standard deviation of gaussian variable

# Plots
matplotlib.pyplot.plot(gaussian_var)
matplotlib.pyplot.show()

# Simple correlation plot
matplotlib.pyplot.scatter(gaussian_var, gaussian_var2) # uncorrelated variables
matplotlib.pyplot.axis('equal') # Equal scaling of axis
matplotlib.pyplot.show()

# multivariate
matplotlib.pyplot.scatter(gaussian_multivar[0,:], gaussian_multivar[0,:]) # variable with itself, i.e., correlation 1

matplotlib.pyplot.scatter(gaussian_multivar[0,:], gaussian_multivar[1,:]) # variable with second variable, i.e., correlation 0

matplotlib.pyplot.scatter(gaussian_multivar[0,:], gaussian_multivar[2,:]) # variable with third, i.e., correlation 0.9

matplotlib.pyplot.scatter(gaussian_multivar[0,:], gaussian_multivar[3,:]) # variable with fourth, i.e., correlation 0.3

matplotlib.pyplot.axis('equal') # Equal scaling of axis
matplotlib.pyplot.show()

# Histograms
matplotlib.pyplot.hist(gaussian_var)
matplotlib.pyplot.hist(gaussian_var2)
matplotlib.pyplot.show()

matplotlib.pyplot.hist(gaussian_multivar.T)
matplotlib.pyplot.show()
```

Some mixing of two random variables

```python
# linear model equal importance
gauss_lin = 1*gaussian_var + 1*gaussian_var2

# Compute and write correlation coefficient
print(numpy.corrcoef(gauss_lin,gaussian_var))
print(numpy.corrcoef(gauss_lin,gaussian_var2))

# Simple correlation plot
matplotlib.pyplot.scatter(gauss_lin, gaussian_var) 
matplotlib.pyplot.scatter(gauss_lin, gaussian_var2)

matplotlib.pyplot.axis('equal') # Equal scaling of axis
matplotlib.pyplot.show()


# linear model one is more important
gauss_lin2 = 10*gaussian_var + 1*gaussian_var2

# Compute and write correlation coefficient
print(numpy.corrcoef(gauss_lin2,gaussian_var))
print(numpy.corrcoef(gauss_lin2,gaussian_var2))

# Simple correlation plot
matplotlib.pyplot.scatter(gauss_lin2, gaussian_var) 
matplotlib.pyplot.scatter(gauss_lin2, gaussian_var2)
matplotlib.pyplot.axis('equal') # Equal scaling of axis
matplotlib.pyplot.show()

# simple nonlinear model (squaring of sum)
gauss_nlin = (1*gaussian_var + 1*gaussian_var2)**2

# Compute and write correlation coefficient
print(numpy.corrcoef(gauss_nlin,gaussian_var))
print(numpy.corrcoef(gauss_nlin,gaussian_var2))

# Simple correlation plot
matplotlib.pyplot.scatter(gauss_nlin, gaussian_var) 
matplotlib.pyplot.scatter(gauss_nlin, gaussian_var2)
matplotlib.pyplot.axis('equal') # Equal scaling of axis
matplotlib.pyplot.show()

# simple nonlinear model (squaring of sum), but one input is less important
gauss_nlin2 = (10*gaussian_var + 1*gaussian_var2)**2

print(numpy.corrcoef(gauss_nlin2,gaussian_var))
print(numpy.corrcoef(gauss_nlin2,gaussian_var2))

# Simple correlation plot
matplotlib.pyplot.scatter(gauss_nlin2, gaussian_var) 
matplotlib.pyplot.scatter(gauss_nlin2, gaussian_var2)
matplotlib.pyplot.axis('equal') # Equal scaling of axis
matplotlib.pyplot.show()
```
