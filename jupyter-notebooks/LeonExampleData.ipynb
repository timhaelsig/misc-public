Import Libraries that are necessary

```python
import numpy
import matplotlib
import matplotlib.pyplot

import pandas
```

Define or import some variables

```python
# Load values from table
patient_data = pandas.read_excel('CVrisk_test_02042020.xlsx', index_col=0, header=2)

print(patient_data.T)
#print(type(patient_data))
#print(patient_data.transpose)

corr_mtx = numpy.corrcoef(patient_data.T)

print(type(corr_mtx))
print(corr_mtx.shape)

matplotlib.pyplot.imshow(corr_mtx);
matplotlib.pyplot.colorbar()
matplotlib.pyplot.show()
```

Do something with the values

```python
patient_data.T
```

```python
# Oldest and youngest
print(numpy.max(patient_data['Alter (Jahre)']))
print(numpy.min(patient_data['Alter (Jahre)']))

# Mean and median age
print(numpy.mean(patient_data['Alter (Jahre)']))
print(numpy.median(patient_data['Alter (Jahre)']))

# Plot age distribution
matplotlib.pyplot.hist(patient_data['Alter (Jahre)'])
matplotlib.pyplot.title('Altersverteilung')
matplotlib.pyplot.xlabel('Alter (Jahre)')
matplotlib.pyplot.ylabel('Häufigkeit in den Daten')
matplotlib.pyplot.grid(True)
matplotlib.pyplot.show()

# Plot cholesterol distribution
matplotlib.pyplot.hist(patient_data['LDL-Cholesterol'], edgecolor='white')
matplotlib.pyplot.show()
```

Simple Correlation Plots

```python
# Plot age vs Cholesterol
matplotlib.pyplot.scatter(patient_data['Alter (Jahre)'], patient_data['LDL-Cholesterol']) 

matplotlib.pyplot.axis('equal') # Equal scaling of axis
matplotlib.pyplot.show()

# Plot age vs several variables
matplotlib.pyplot.scatter(patient_data['Alter (Jahre)'], patient_data['Nüchtern Blutzucker (mg/dl)']) 
matplotlib.pyplot.scatter(patient_data['Alter (Jahre)'], patient_data['24h-MAP (mmHg)']) 

matplotlib.pyplot.axis('equal') # Equal scaling of axis
matplotlib.pyplot.show()
```
